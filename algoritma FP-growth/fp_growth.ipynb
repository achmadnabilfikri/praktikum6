```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
```

```python
transactions = [
    ["milk", "flour"], ["eggs", "rice", "butter", "milk", "flour"], ["sugar", "cheese"], ["flour"], 
    ["milk", "flour"], ["eggs", "rice", "butter", "cheese"], ["eggs", "rice", "butter", "milk", "flour"],
    ["sugar", "cheese"], ["milk", "sugar", "flour", "cheese"], ["sugar", "cheese"], ["flour"],
    ["eggs", "rice", "butter", "sugar", "cheese"], ["sugar", "milk"], ["milk", "flour"],
    ["butter", "eggs"], ["milk", "sugar", "cheese"], ["milk", "flour", "sugar"],
    ["eggs", "milk", "flour", "rice", "butter",], ["milk", "flour"], ["milk"], ["sugar", "butter"],
    ["rice", "eggs"], ["eggs", "rice", "butter", "milk", "flour"], ["sugar", "milk"],
    ["flour", "sugar", "butter", "milk", "cheese"],["eggs", "flour", "rice", "butter",]]
```

```python
my_transactionencoder = TransactionEncoder()
my_transactionencoder.fit(transactions)
encoded_transactions = my_transactionencoder.transform(transactions)
encoded_transactions_df = pd.DataFrame(encoded_transactions, columns=my_transactionencoder.columns_)

encoded_transactions_df
```

```python
min_support = 7/len(transactions) 

from mlxtend.frequent_patterns.fpgrowth import fpgrowth
frequent_itemsets = fpgrowth(encoded_transactions_df, min_support=min_support, use_colnames = True)

frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
```
